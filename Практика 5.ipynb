<p>Реализация планировщика с тремя состояниями:</p>
<p>ГОТОВ <> ВЫПОЛНЕНИЕ > ОЖИДАНИЕ</p>

```python
import threading
from time import sleep

def thread(done, name):
    global k
    for i in range(1,11):
        sleep(float(k[name-1]))
        done.wait()
        with lock:
            print('Поток',name,': ',i,'; приоритет:', int(k[name-1]))

q = 2
l = 10
kk = 10
k = [4, 4, 4]
lock = threading.Lock()
done1 = threading.Event()
done2 = threading.Event()
done3 = threading.Event()

done1.set()
done2.set()
done3.set()

thread1 = threading.Thread(target=thread, args=[done1, 1], daemon=True)
thread2 = threading.Thread(target=thread, args=[done2, 2], daemon=True)
thread3 = threading.Thread(target=thread, args=[done3, 3], daemon=True)

thread1.start()
thread2.start()
thread3.start()

while kk != "5":
    kk = input('Контроль/Приоритет: ')
    if kk != "5"
    l = input('Поток: ')
    q = input('Действие: ')
    if kk == "1":
        if q == "0":
            if l == "1":
                done1.clear()
            if l == "2":
                done2.clear()
            if l == "3":
                done3.clear()

        if q == "1":
            if l == "1":
                done1.set()
            if l == "2":
                done2.set()
            if l == "3":
                done3.set()
    if kk == "2":
        if q == "0":
            if l == "1":
                k[0] = k[0]*2
            if l == "2":
                k[0] = k[0]*2
            if l == "3":
                k[0] = k[0]*2

        if q == "1":
            if l == "1":
                k[0] = k[0]/2
            if l == "2":
                k[1] = k[1]/2
            if l == "3":
                k[2] = k[2]/2

        
done1.set()
done2.set()
done3.set()
```
